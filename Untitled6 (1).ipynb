**preprocessing**

```python
data_encoded = pd.get_dummies(data, columns=['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course'])
data_cleaned = data_encoded.dropna()

```

```python
# Separate features (x) and target variable (y)
x_reg = data_cleaned.drop(['math score'], axis=1)
y_reg = data_cleaned['math score']
```

```python
# Split the dataset into training and testing sets
x_reg_train, x_reg_test, y_reg_train, y_reg_test = train_test_split(x_reg, y_reg, test_size=0.2, random_state=0)
```

**Linear Regression Model**

```python
reg_model = LinearRegression()
reg_model.fit(x_reg_train, y_reg_train)
```

```python
# Make Predictions on Testing Set
reg_predictions = reg_model.predict(x_reg_test)
```

```python
# Calculate Mean Squared Error
reg_mse = mean_squared_error(y_reg_test, reg_predictions)

print(f'Regression Mean Squared Error: {reg_mse}')
```

**Random Forest Regressor Model**

```python
data['above_threshold'] = (data['math score'] > 70).astype(int)

```

```python

data_encoded = pd.get_dummies(data, columns=['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course'])

```

```python
x_cls = data_encoded.drop(['math score', 'above_threshold'], axis=1)
y_cls = data_encoded['above_threshold']
```

```python
# Split the dataset into training and testing sets
x_cls_train, x_cls_test, y_cls_train, y_cls_test = train_test_split(x_cls, y_cls, test_size=0.2, random_state=0)
```

```python
cls_model = RandomForestClassifier(n_estimators=100, random_state=0)
cls_model.fit(x_cls_train, y_cls_train)
```

```python
# Make Predictions on Testing Set
cls_predictions = cls_model.predict(x_cls_test)
```

```python
# Calculate Accuracy
cls_accuracy = accuracy_score(y_cls_test, cls_predictions)

print(f'Classification Accuracy: {cls_accuracy}')
```

```python
conf_matrix = confusion_matrix(y_cls_test, cls_predictions)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Class')
plt.ylabel('Actual Class')
plt.title('Confusion Matrix')
plt.show()
```

```python
comparison_df = pd.DataFrame({'Actual': y_cls_test, 'Predicted': cls_predictions})
comparison_df['Correct'] = comparison_df['Actual'] == comparison_df['Predicted']
print('\nComparison Table:')
print(comparison_df)
```
